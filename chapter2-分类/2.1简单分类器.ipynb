```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = np.array([
    [3, 1], [2, 5], [1, 8], [6, 4], [5, 2], [3, 5], [4, 7], [4, -1]
])
y = [0, 1, 1, 0, 0, 1, 1, 0]
```

```python
# 分类
class_0 = np.array([X[i] for i in range(len(X)) if y[i] == 0])
class_1 = np.array([X[i] for i in range(len(X)) if y[i] == 1]) 

# 绘制点图
plt.figure()
plt.scatter(class_0[:, 0], class_0[:, 1], color="green", marker='s')
plt.scatter(class_1[:, 0], class_1[:, 1], color="green", marker='x')
plt.show()
```

```python
#可以使用简单的线性分类
line_x = range(10)
line_y = line_x

# 绘制点图
plt.figure()
plt.scatter(class_0[:, 0], class_0[:, 1], color="green", marker='s')
plt.scatter(class_1[:, 0], class_1[:, 1], color="green", marker='x')
plt.plot(line_x, line_y, color="black", linewidth=4)
plt.show()
```
